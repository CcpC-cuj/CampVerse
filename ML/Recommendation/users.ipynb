```python
import pandas as pd
import pickle
```

```python
users= pd.read_json('Datasets/updated_users.json')
```

```python
events = pickle.load(open('Models/events.pkl', 'rb'))
```

```python
users.head()
```

```python
users['events'] = users['explored_history']+users['liked_events']+users['attended_events']
```

```python
users.head(1)
```

```python
users['key']= users['preferred_domains'] + users['interests_tags']
```

```python
users['key']=users['key'].apply(lambda x:" ".join(x))
```

```python
users['key']=users['key']+users['location']
```

```python
users = users[['user_id','name','key', 'events']]
```

```python
users
```

```python
events.head()
```

```python
#adding events key to the user event column
#finding event key by event id
# 1. Explode the liked_events column
exploded = users.explode('events')

# 2. Map event_ids to event_names
id_to_key = dict(zip(events['event_id'], events['key']))
exploded['events'] = exploded['events'].map(id_to_key)

# 3. Group back to list per user
users['events'] = exploded.groupby(level=0)['events'].apply(list)
```

```python
users.head(2)
```

```python
users['events']= users['events'].apply(lambda x:" ".join(x))
```

```python
users.head(2)
```

```python
users['key'] = users['key'] + users['events']
```

```python
users.drop(columns=['events'], inplace=True)
```

```python
users.head(1)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
user_vector = cv.fit_transform(users['key']).toarray()
```

```python
user_vector[2]
```

```python
user_similarity = cosine_similarity(user_vector)
```

```python
list(enumerate(user_similarity[0]))
```

```python
def recommend_user(user_id):
    index = users[users['user_id']== user_id].index[0]
    distances = user_similarity[index]
    users_list = sorted(list(enumerate(distances)),reverse=True , key=lambda x:x[1])[1:6]

    for i in users_list:
        print(users.iloc[i[0]].user_id)

    return
```

```python
recommend_user('u320')
```
