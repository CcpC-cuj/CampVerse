```python
users['events']= users['events'].apply(lambda x:" ".join(x))
```

```python
users['key'] = users['key'] + " " + users['events'].apply(lambda x: " ".join(x))
```

```python
users.head(1)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
user_vector = cv.fit_transform(users['key']).toarray()
```

```python
user_vector[2]
```

```python
user_similarity = cosine_similarity(user_vector)
```

```python
user_similarity.shape
```

```python
list(enumerate(user_similarity[0]))
```

```python
def recommend_user(user_id):
    index = users[users['user_id']== user_id].index[0]
    distances = user_similarity[index]
    users_list = sorted(list(enumerate(distances)),reverse=True , key=lambda x:x[1])[1:6]

    for i in users_list:
        print(users.iloc[i[0]].user_id)

    return
```

```python
recommend_user('68e0e01c6fbdb8a90b76129b')
```
